# Approximating $\pi$ using Monte Carlos

```python
# IMPORTS
import numpy as np
import matplotlib.pyplot as plt
```

Given the area of a circle is $A_c=\pi \cdot r^2$ and the area of a square is $A_s = l^2$, then we can approximate $\pi$ as follows: $\frac{A_c}{A_s} = \frac{\pi \cdot r^2}{l^2}$. If $l = 2r$, then $\frac{A_c}{A_s} = \frac{\pi \cdot r^2}{(2r)^2} = \frac{\pi \cdot r^2}{4r^2} = \frac{\pi}{4} $. Then, $\pi = 4 \cdot \frac{A_c}{A_s} $

### Using Monte Carlos

The circle will be placed at the center of the square and a set of randomly generated coordinate points will be placed in the domain and range of the square. As the number of generated coordinate points increase, the approximation of pi will become more accurate.  

Define the radius of a circle. Use this radius as the length of the square

### Code

```python
# RADIUS OF CIRCLE
rad = 20

# LENGTH OF SQUARE
sqLen = 2*rad

# CENTER OF CIRCLE
cirCenter = np.array([sqLen/2,sqLen/2])
```

Define the number of randomly generated points that will be used.

```python
# NUMBER OF RANDOM POINTS
nPoints = 1000

# ARRAY WITH (X,Y) COORDINATE POINTS 
randCoor = np.random.rand(nPoints , 2)*sqLen
```

Determine the number of points that lie inside the circle

```python
logicalArr = (randCoor[:,0] - cirCenter[0])**2 + (randCoor[:,1] - cirCenter[1])**2 <= rad**2
nCoorInsideCircle = np.sum(logicalArr)
```

Determine the proportion of the # of coordinate points inside over # of coordinate points inside the square ( number of points)

```python
approxPi = 4 * (nCoorInsideCircle / nPoints)

print('Given {} Coordinate Points, Approx. Pi value is:{}'.format(nPoints,approxPi))
```

```python
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(1,1,1)

# PLOT THE GENERATED POINTS
ax1.scatter(randCoor[:,0],randCoor[:,1], s = 2)

# DRAW A CIRCLE WITH RADIUS 'rad' WITH A CENTER AT '(cirCenter[0], cirCenter[0])'
circle1  = plt.Circle((cirCenter[0], cirCenter[0]), rad , color='red', fill = False)

# PLACE CIRCLE INTO THE AX1
ax1.add_artist(circle1)

# SET THE LIMITS TO THE SIZE OF THE SQUARE
ax1.set_xlim([0,sqLen])
ax1.set_ylim([0,sqLen])

# ENSURE THAT THE AX HAS THE SAME LENGTH ON XY AXIS
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
